##### Kernel Density Plot (KDE)
A `Kernel Density Estimation (KDE)` plot is a technique used to visualize the probability distribution of a continuous variable. It serves as a smooth alternative to a histogram and is often used to estimate a variable's probability density function. KDE plots are helpful for revealing the distribution's shape and identifying modes (peaks) in the data.

`Probability density` refers to the likelihood that a continuous random variable will assume a specific value. It is a key concept in both statistics and probability theory.

##### Key Concepts
- `Kernel`: The kernel is a weighting function used in density estimation. Popular choices include the Gaussian (normal) kernel, Epanechnikov, and uniform kernels.
  
- `Bandwidth`: Bandwidth controls the smoothness of the KDE curve by adjusting the width of the kernel. A larger bandwidth results in a smoother density curve, while a smaller bandwidth captures finer details but may introduce noise.

- `Density Estimation`: This process involves placing a kernel (such as a Gaussian) at each data point and summing the kernels to estimate the overall density at various points in the dataset.

# Applying dimensionality reduction
Converting the DataFrame into 2D DataFrame for visualization

```python
pca = PCA(n_components=2)
principal_components = pca.fit_transform(scaled_df)
pca_df = pd.DataFrame(data=principal_components ,columns=["PCA1","PCA2"])
pca_df
```

# Hyperparameter tuning
## Finding 'k' value by Elbow Method

```python
inertia = []
range_val = range(1,15)
for i in range_val:
    kmean = KMeans(n_clusters=i)
    kmean.fit_predict(pd.DataFrame(scaled_df))
    inertia.append(kmean.inertia_)
plt.plot(range_val,inertia,'bx-')
plt.xlabel('Values of K') 
plt.ylabel('Inertia') 
plt.title('The Elbow Method using Inertia') 
plt.show()
```

From the graph we can see that the elbow started forming at K = 4

# Model Building using KMeans

```python

kmeans_model=KMeans(4)
kmeans_model.fit_predict(scaled_df)
pca_df_kmeans= pd.concat([pca_df,pd.DataFrame({'cluster':kmeans_model.labels_})],axis=1)
```

# Visualizing the clustered dataframe

```python
plt.figure(figsize=(8,8))
ax=sns.scatterplot(x="PCA1",y="PCA2",hue="cluster",data=pca_df_kmeans,palette=['red','green','blue','black'])
plt.title("Clustering using K-Means Algorithm")
plt.show()
```

*`When you use an elbow method to find your K its good to try numbers that are before and after the k-value you got just to be sure that your judgement is right.`*

```python
# find all cluster centers
cluster_centers = pd.DataFrame(data=kmeans_model.cluster_centers_,columns=[df.columns])
# inverse transform the data
cluster_centers = scalar.inverse_transform(cluster_centers)
cluster_centers = pd.DataFrame(data=cluster_centers,columns=[df.columns])
cluster_centers
```

# Training and Testing the model accuracy using decision tree

```python
#Split Dataset
X = cluster_df.drop(['Cluster'],axis=1)
y= cluster_df[['Cluster']]
X_train, X_test, y_train, y_test =train_test_split(X, y, test_size=0.3)
```

```python
X_train
```

```python
X_test
```

```python
#Decision_Tree
model= DecisionTreeClassifier(criterion="entropy")
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

```python
#Confusion_Matrix
print(metrics.confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```
