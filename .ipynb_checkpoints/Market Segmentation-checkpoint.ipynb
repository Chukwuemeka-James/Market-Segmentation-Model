
##### Kernel Density plot(kde)
A `Kernel Density` Plot (KDE Plot) is a method for visualizing the probability density of a continuous variable. It is a smoothed version of the histogram and is used to estimate the probability density function of a variable. KDE plots are particularly useful for identifying the shape of the data distribution and highlighting modes (peaks) in the data.

`Probability density` is a concept in statistics and probability theory that describes the likelihood of a continuous random variable taking on a particular value. 

##### Key Concepts
`Kernel`: A kernel is a weighting function used in the estimation. Common kernels include Gaussian (normal), Epanechnikov, and uniform.

`Bandwidth`: The bandwidth is a smoothing parameter that determines the width of the kernel and thus controls the smoothness of the resulting density curve. A larger bandwidth leads to a smoother curve, while a smaller bandwidth captures more detail but can also introduce noise.

`Density Estimation`: The process involves placing a kernel (e.g., Gaussian) on each data point and summing these kernels to get the density estimate at each point. point.

### K-means Clustering [Clustering, Centroid and Classification]

`K-means clustering` is a method used to group data points into a specified number (K) of clusters. The goal is to have data points in the same cluster be more similar to each other than to those in other clusters. `K` in this case is the number of clusters or groups we ask the algorithm to create.

##### Clustering
In simple terms, clustering and classification are two different types of machine learning tasks that deal with grouping data, but they do so in different ways and with different goals:

##### Clustering
1. Unsupervised Learning: Clustering is an unsupervised learning task. This means that it doesn't require labeled data to train the model. The algorithm tries to find patterns and group the data based on similarities or distances between data points.
2. Group Discovery: The goal of clustering is to discover the inherent groupings or clusters in the data. You don't tell the algorithm how many or what kinds of groups to find; it figures this out on its own.
3. No Labels Needed: Since clustering doesn't use labels, it can't tell you what category a new data point belongs to. It only tells you which cluster a data point is similar to based on the learned structure.
    `Example:` Grouping customers based on their purchasing behavior to identify different segments or groups of customers with similar behaviors.

##### Classification
1. Supervised Learning**: Classification is a supervised learning task. This means it requires labeled data to train the model. The algorithm learns from examples where the input data is paired with the correct output label.
2. Predictive Modeling: The goal of classification is to predict the label or category of new, unseen data points. The algorithm uses the learned patterns from the labeled training data to make predictions.
3. Labels Required: Since classification uses labels, it can provide a specific category or class for new data points.
    `Example:` Classifying emails as either "spam" or "not spam" based on examples of labeled emails.

*These distinctions make `clustering` suitable for tasks where the goal is to explore and understand the structure of the data, and `classification` suitable for tasks where the goal is to categorize data into predefined labels.*

##### Centroid
`Center Point:` The centroid is the central point of a cluster in a dataset. It is calculated as the mean (average) position of all the points in the cluster. It serves as a representative or summary of the entire cluster, capturing the central location of the data points within that cluster.
*In clustering algorithms like K-means, centroids are used to define the clusters. The algorithm iteratively adjusts the centroids to better fit the data points.*

# Hyperparameter tuning
## Finding 'k' value by Elbow Method

```python
inertia = []
range_val = range(1,15)
for i in range_val:
    kmean = KMeans(n_clusters=i)
    kmean.fit_predict(pd.DataFrame(scaled_df))
    inertia.append(kmean.inertia_)
plt.plot(range_val,inertia,'bx-')
plt.xlabel('Values of K') 
plt.ylabel('Inertia') 
plt.title('The Elbow Method using Inertia') 
plt.show()
```

From the graph we can see that the elbow started forming at K = 4

# Model Building using KMeans

```python

kmeans_model=KMeans(4)
kmeans_model.fit_predict(scaled_df)
pca_df_kmeans= pd.concat([pca_df,pd.DataFrame({'cluster':kmeans_model.labels_})],axis=1)
```

# Visualizing the clustered dataframe

```python
plt.figure(figsize=(8,8))
ax=sns.scatterplot(x="PCA1",y="PCA2",hue="cluster",data=pca_df_kmeans,palette=['red','green','blue','black'])
plt.title("Clustering using K-Means Algorithm")
plt.show()
```

*`When you use an elbow method to find your K its good to try numbers that are before and after the k-value you got just to be sure that your judgement is right.`*

```python
# find all cluster centers
cluster_centers = pd.DataFrame(data=kmeans_model.cluster_centers_,columns=[df.columns])
# inverse transform the data
cluster_centers = scalar.inverse_transform(cluster_centers)
cluster_centers = pd.DataFrame(data=cluster_centers,columns=[df.columns])
cluster_centers
```

# Training and Testing the model accuracy using decision tree

```python
#Split Dataset
X = cluster_df.drop(['Cluster'],axis=1)
y= cluster_df[['Cluster']]
X_train, X_test, y_train, y_test =train_test_split(X, y, test_size=0.3)
```

```python
X_train
```

```python
X_test
```

```python
#Decision_Tree
model= DecisionTreeClassifier(criterion="entropy")
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

```python
#Confusion_Matrix
print(metrics.confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```
